```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

Zadanie 3

```python
function = lambda x: x**2 + 5
args_1 = np.linspace(-1, 1)
args_2 = np.linspace(-6, 6)
args_3 = np.linspace(0, 5)

plt.plot(args_1, function(args_1), label="y = x^2 + 5")
plt.xlabel("arguments")
plt.ylabel("values")
plt.title('x>-1, x<1')
plt.legend()
plt.grid()
plt.show()

plt.plot(args_2, function(args_2), label="y = x^2 + 5")
plt.xlabel("arguments")
plt.ylabel("values")
plt.title('x>-6, x<6')
plt.legend()
plt.grid()
plt.show()

plt.plot(args_3, function(args_3), label="y = x^2 + 5")
plt.xlabel("arguments")
plt.ylabel("values")
plt.title('x>0, x<5')
plt.legend()
plt.grid()
plt.show()
```
