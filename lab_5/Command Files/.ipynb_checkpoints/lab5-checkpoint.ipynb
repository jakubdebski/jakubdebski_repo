```python
import pandas as pd
import matplotlib.pyplot as plt
directory = "../Analysis Data/"
df = pd.read_csv(f"{directory}11_S╠üLA╠ĘSKIE.csv")
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
display(df)
```

```python
fig, ax = plt.subplots(1,1)
ax.hist(df.iloc[:,0], bins='auto', rwidth=0.85)
ax.set_xlabel(df.columns[0])
ax.set_ylabel('Ilość')
plt.grid()
plt.show()
```
